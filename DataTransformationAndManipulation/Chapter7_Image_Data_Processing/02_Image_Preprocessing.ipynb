## Part 2: Image Preprocessing

In this notebook, we will explore various techniques for preprocessing image data. Preprocessing is an essential step in image analysis and machine learning, as it helps in standardizing the input data, making it suitable for further analysis and model training. The preprocessing techniques we will cover include resizing and cropping, converting to grayscale, and pixel normalization.

### 2.1 Resizing and Cropping

Resizing and cropping are common preprocessing steps that adjust the dimensions of an image. Resizing scales the image to a specified size, while cropping removes unwanted parts of the image, focusing on the region of interest.

### 2.2 Converting to Grayscale

Converting an image to grayscale reduces its complexity by removing color information. This is useful for tasks where color is not a crucial feature, simplifying the analysis and reducing computational load.

### 2.3 Pixel Normalization

Pixel normalization scales the pixel values of an image to a standard range, typically between 0 and 1 or -1 and 1. This step is important for ensuring that the input data has a consistent scale, improving the performance and stability of machine learning models.


```python
# Import necessary libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Function to display images
def display_image(image, title='Image', cmap=None):
    plt.figure(figsize=(6, 6))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load an example image
image_path = 'image.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB for correct color display
```

```python
# Display original image
display_image(image, title='Original Image')
```

```python
# Resizing the image
resized_image = cv2.resize(image, (128, 128))
display_image(resized_image, title='Resized Image')
```

```python
# Cropping the image (center crop)
h, w, _ = image.shape
start_x = w//4
start_y = h//4
end_x = start_x + w//2
end_y = start_y + h//2
cropped_image = image[start_y:end_y, start_x:end_x]
display_image(cropped_image, title='Cropped Image')
```

```python
# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
display_image(grayscale_image, title='Grayscale Image', cmap='gray')
```

```python
# Normalize pixel values to the range [0, 1]
normalized_image = image / 255.0
display_image(np.clip(normalized_image, 0, 1), title='Normalized Image [0, 1]')
```

```python
# Normalize pixel values to the range [-1, 1]
normalized_image_2 = (image / 127.5) - 1
display_image(np.clip((normalized_image_2 + 1) / 2, 0, 1), title='Normalized Image [-1, 1]')
```
