# Principal Component Analysis (PCA)

In this notebook, we will cover the concept of PCA. We will discuss the context of dimensionality reduction and the principal component analysis.

## Dimensionality Reduction

### What is Dimensionality Reduction?
- A set of techniques to reduce the size of a dataset.
- Can consider features, records, or other information for reduction.
- Can be seen as a compression technique.
- Often used as a preliminary step before using a machine learning algorithm.

### Use Cases:
- Presence of a large number of features in the dataset.
- Correlation between a certain number of features.
- Features not appropriate for the problem at hand.
- Visualization: Too many features (more than 3) making visualization impossible.
  - More appropriate in this case to project onto a 2D plane.

### Constraint:
- The representation of the data changes, but the information contained should not change dramatically.
- The goal is to simplify the data representation while retaining maximum information.

### Dimensionality Reduction Techniques:
1. Feature Selection: Selecting a set of features based on criteria.
2. Feature Extraction: Creating a new set of coordinates that simplify the data representation.

## Principal Component Analysis (PCA)

### PCA:
- A dimensionality reduction technique that can be used on a matrix of any dimension.
- PCA produces a new coordinate system where each component retains as much variance as in the original data.
- PCA of a matrix A is the eigenvalue decomposition of the covariance matrix of A.
- It's the direction where there is the most variance, where there is the most dispersion of data.

### Eigenvector and Eigenvalue
- Eigenvectors and eigenvalues come in pairs: each eigenvector has an eigenvalue.
- **Eigenvector**: The direction in which we have the maximum dispersion.
- **Eigenvalue**: The value of the dispersion in the direction of the eigenvector.
- The number of eigenvectors/eigenvalues is equal to the number of dimensions in the dataset.
  - For example, if we have age, weight, and height, we will have 3 eigenvectors/eigenvalues.
- The principal component is the eigenvector with the maximum eigenvalue.

### PCA Objective
- PCA seeks the dimensions in which the maximum variance occurs.
- It's a transformation to another coordinate system ordered by variance.

### Variance:
- The average distance from the mean of a dataset to a point in the dataset.
- Provides an idea of data spread.

### Covariance:
- Measures similar to variance but in multiple dimensions.
- Assesses the change in two variables taken together relative to each other.

### Covariance Matrix:
- Given matrix X, we can define a covariance matrix.
- Elements on the diagonal are the variances of $X_i$.
- The covariance matrix is square.

### Eigenvalue Decomposition:
- For each square matrix A, we have 
> $$A = Q \Lambda Q^{-1}$$
- Columns of Q: eigenvectors of A.
- Values in $\Lambda$: eigenvalues of A.

### PCA Transformation
- We seek to transform the base representation.
- We proceed by ordering according to the magnitude of the variance.

### Example: Implementing PCA in Python


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Generate fictitious data with 5 dimensions
np.random.seed(42)
X = np.random.rand(100, 5) * 10  # 5-dimensional data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame for PCA results
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])

# Plot the results
plt.scatter(df_pca['Principal Component 1'], df_pca['Principal Component 2'], color='blue')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Results')
plt.show()

# Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)

# Original DataFrame for reference
df_original = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])
print(df_original.head())

# PCA DataFrame
print(df_pca.head())
```
