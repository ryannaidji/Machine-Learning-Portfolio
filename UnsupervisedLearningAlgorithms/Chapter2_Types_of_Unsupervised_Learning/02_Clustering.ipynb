# Clustering Methods

Clustering is a fundamental task in unsupervised learning, which involves grouping a set of objects in such a way that objects in the same group (called a cluster) are more similar to each other than to those in other groups (clusters). Clustering is widely used in various fields, such as data mining, pattern recognition, image analysis, and bioinformatics.

In this notebook, we will provide an overview of different clustering methods, which will be explored in more detail in subsequent chapters.

## Types of Clustering Methods

1. **Centroid-Based Clustering**
   - K-Means
   - K-Medoids
2. **Connectivity-Based Clustering (Hierarchical Clustering)**
   - Agglomerative Clustering
   - Divisive Clustering
3. **Distribution-Based Clustering**
   - Gaussian Mixture Models (GMM)
4. **Density-Based Clustering**
   - DBSCAN

Each clustering method has its own strengths and weaknesses and is suitable for different types of data and applications.


## Centroid-Based Clustering

Centroid-based clustering organizes the data into non-hierarchical clusters. In this method, clusters are represented by a central vector, which is not necessarily a member of the data set. The most popular centroid-based clustering algorithm is K-Means.

### K-Means Clustering
K-Means clustering aims to partition `n` observations into `k` clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster.


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

## Connectivity-Based Clustering (Hierarchical Clustering)

Connectivity-based clustering, also known as hierarchical clustering, is a method of cluster analysis that seeks to build a hierarchy of clusters. There are two main types of hierarchical clustering:

1. **Agglomerative Clustering**: This is a bottom-up approach where each observation starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy.
2. **Divisive Clustering**: This is a top-down approach where all observations start in one cluster, and splits are performed recursively as one moves down the hierarchy.


```python
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=100, centers=3, random_state=42)

# Apply hierarchical clustering
Z = linkage(X, 'ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

## Distribution-Based Clustering

Distribution-based clustering is based on the concept that data points are generated by a mixture of probability distributions. The most commonly used distribution-based clustering method is Gaussian Mixture Models (GMM).

### Gaussian Mixture Models (GMM)
Gaussian Mixture Models assume that the data is a mixture of several Gaussian distributions with unknown parameters. GMM uses the Expectation-Maximization (EM) algorithm to estimate the parameters.


```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply GMM clustering
gmm = GaussianMixture(n_components=4)
gmm.fit(X)
y_gmm = gmm.predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, s=50, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

## Density-Based Clustering

Density-based clustering is a method that finds clusters based on the density of data points in a region. One of the most popular density-based clustering algorithms is DBSCAN (Density-Based Spatial Clustering of Applications with Noise).

### DBSCAN
DBSCAN groups together points that are closely packed together, marking as outliers the points that lie alone in low-density regions.


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```

## Why Use Clustering?

Clustering is a powerful tool in unsupervised learning that helps identify patterns and structures in data without prior knowledge of the labels. It is used for various reasons:

1. **Exploratory Data Analysis (EDA)**:
   - Clustering helps to explore the underlying structure of the data.
   - It provides insights into the natural grouping of data points.

2. **Data Reduction**:
   - Clustering can reduce the size of large datasets by categorizing similar data points.
   - It simplifies data for further analysis.

3. **Anomaly Detection**:
   - Clustering can identify outliers or anomalies that do not fit into any cluster.
   - Useful in fraud detection, network security, and other applications.

4. **Segmentation**:
   - Clustering is used to segment data into different groups based on similarity.
   - Commonly used in market segmentation, image segmentation, and customer segmentation.

5. **Feature Engineering**:
   - Clustering results can be used as features for other machine learning algorithms.
   - Enhances the predictive power of models by incorporating cluster-based features.

6. **Recommendation Systems**:
   - Clustering helps to group users or items with similar preferences.
   - Improves the accuracy of recommendation systems.

## When to Use Clustering?

Clustering is particularly useful in the following scenarios:

1. **Unlabeled Data**:
   - When the data lacks predefined labels or categories.
   - Clustering can discover hidden patterns and groupings.

2. **Data Exploration**:
   - During the initial stages of data analysis to gain insights into the data distribution.
   - Helps in identifying the number of natural groups or clusters.

3. **Anomaly Detection**:
   - When the goal is to detect unusual patterns or outliers.
   - Clustering algorithms like DBSCAN are effective for this purpose.

4. **Customer Segmentation**:
   - In marketing to segment customers based on purchasing behavior, preferences, etc.
   - Helps in targeted marketing and personalized recommendations.

## Differences Between Clustering Methods

Different clustering methods have unique characteristics and are suited for various types of data and applications:

### 1. Centroid-Based Clustering (K-Means, K-Medoids)
- **K-Means**:
  - Groups data into `k` clusters based on the nearest mean.
  - Efficient for large datasets.
  - Sensitive to initial centroids and outliers.
- **K-Medoids**:
  - Similar to K-Means but uses medoids (actual data points) as cluster centers.
  - More robust to outliers than K-Means.

### 2. Connectivity-Based Clustering (Hierarchical Clustering)
- **Agglomerative Clustering**:
  - Starts with individual points and merges them into clusters.
  - Creates a dendrogram representing the hierarchy of clusters.
  - Computationally intensive for large datasets.
- **Divisive Clustering**:
  - Starts with a single cluster and recursively splits it.
  - Less common than agglomerative clustering.

### 3. Distribution-Based Clustering (Gaussian Mixture Models - GMM)
- **GMM**:
  - Assumes data is generated from a mixture of Gaussian distributions.
  - Provides probabilistic cluster assignments.
  - Can model complex cluster shapes but may require more computational resources.

### 4. Density-Based Clustering (DBSCAN)
- **DBSCAN**:
  - Groups data points based on density (regions with many data points).
  - Can find arbitrarily shaped clusters and identify outliers.
  - Does not require specifying the number of clusters in advance.
  - Sensitive to the choice of parameters (epsilon and min_samples).

## Summary

Each clustering method has its own strengths and weaknesses, making them suitable for different scenarios:

- **K-Means**: Efficient for large datasets but sensitive to outliers.
- **K-Medoids**: More robust to outliers but computationally intensive.
- **Hierarchical Clustering**: Good for small datasets and creating dendrograms but not scalable.
- **GMM**: Handles complex shapes and probabilistic assignments but is computationally intensive.
- **DBSCAN**: Finds arbitrarily shaped clusters and outliers but requires careful parameter tuning.

Choosing the right clustering method depends on the nature of the data, the specific application, and the desired outcome.
