# 2. K-Means Clustering

## Introduction to K-Means Clustering

### Principle of the K-Means Algorithm

The K-Means algorithm is a popular partitioning method used to divide a dataset into k distinct clusters based on the similarity of the data points.

### How K-Means Works

* Partitioning methods aim to create a single partition of N individuals into C groups.
* There are multiple methods for partitioning data:
  - K-Means method
  - C-Medoids method
  - Dynamic Clouds

The K-Means algorithm uses an iterative refinement technique that gradually improves the quality of the groups based on a similarity index between the different individuals.

### Steps of the K-Means Algorithm

1. Choose k initial centers $G_i$ with $i = 1, 2, ...k$.
2. Assign each individual to the nearest group $G_i$.
3. Recalculate the new centers $G_i$.
4. Repeat steps 2 and 3 until the centers become stable.

### Advantages and Disadvantages

#### Advantages
* The K-Means algorithm is easy to implement.

#### Disadvantages
* The value of k must be chosen a priori.
* Clusters depend on initialization and the chosen distance metric.

## Example: K-Means Clustering with Multiple Clusters

We will now demonstrate the K-Means algorithm using a synthetic dataset with multiple clusters.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
np.random.seed(42)
X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

# Create and fit the K-Means model
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)

# Predict the cluster labels
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Calculate and print the silhouette score
silhouette_avg = silhouette_score(X, y_kmeans)
print(f'Silhouette Score for K-Means Clustering: {silhouette_avg:.2f}')
```

## Interpretation of Results

In this example, we used the K-Means algorithm to cluster the data into 5 groups. The red 'X' markers represent the centroids of each cluster.

### Silhouette Score
The silhouette score for this clustering is {silhouette_avg:.2f}. A higher silhouette score indicates that the clusters are well-separated and the data points are well-clustered. The silhouette score ranges from -1 to 1, where a score closer to 1 indicates better-defined clusters.
