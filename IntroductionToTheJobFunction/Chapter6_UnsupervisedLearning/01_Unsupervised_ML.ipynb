# 1. Introduction: Unsupervised Machine Learning

## Unsupervised Learning

Unlike supervised learning, in the case of unsupervised learning, the output data is not known. We have n individuals described by their feature vector $D_n = \{x_1, x_2, ..., x_n\}$. The system learns by itself to organize the data or to determine structures within the data.

The most common unsupervised learning task is clustering, which involves grouping input data based on their common characteristics.

### Examples of Applications

#### Epidemiology
In epidemiology, researchers always try to find explanatory hypotheses for the outbreak of an epidemic. For example, given a large dataset of liver cancer victims, we want to generate explanatory hypotheses.

- Data scientists will determine the different groups.
- The epidemiologist would then try to associate various explanatory factors, such as geographical origin, genetics, consumption habits, exposure to potentially or actually toxic agents (heavy metals, toxins like aflatoxin, etc.).

## Clustering

### Principle
Clustering, also known as aggregation, is a statistical method for unsupervised analysis and classification of data.

**Objective**: To form groups or clusters of similar objects from a heterogeneous set of objects.

### Applications

#### Marketing
Market segmentation: Dividing a market into distinct groups of customers based on purchase databases.

#### Environment
Determining similar land zones (for their use) in an earth observation database.

#### Insurance
Identifying distinct groups of insured individuals who have filed a large number of claims.

## Clustering Approaches
Several approaches are used, including:

- **Partitioning**: Constructing several partitions and evaluating them according to certain criteria.
- **Hierarchical**: Creating a hierarchical decomposition of individuals based on certain criteria.
- **Density-based**: Based on notions of connectivity and density.
- Etc.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
np.random.seed(42)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Create and fit the K-Means model
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)

# Predict the cluster labels
y_kmeans = kmeans.predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Calculate and print the silhouette score
silhouette_avg = silhouette_score(X, y_kmeans)
print(f'Silhouette Score for K-Means Clustering: {silhouette_avg:.2f}')
```

## Hierarchical Clustering (Agglomerative Clustering)

```python
# Import necessary libraries for Hierarchical Clustering
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Create and fit the Agglomerative Clustering model
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_hc, s=50, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Plot the dendrogram
linked = linkage(X, 'ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
