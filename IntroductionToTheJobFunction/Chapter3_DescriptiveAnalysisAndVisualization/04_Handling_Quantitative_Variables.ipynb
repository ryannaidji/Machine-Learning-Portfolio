# 4. Handling Quantitative Variables

## 4.1 Description of Quantitative Variables

We distinguish three main families of characteristics:
- **Central Tendency Characteristics**: Indicate the order of magnitude of the data and their central value, i.e., the position around which these values gather.
- **Dispersion Characteristics**: Evaluate the variability of the data and summarize the deviation of all individuals from their central tendency.
- **Shape Characteristics**: Many physical phenomena are distinguished by quantitative variables that follow a normal distribution. The main shape characteristics are the skewness coefficient and Kurtosis.

### Characteristics Summary Table

| Central Tendency Characteristics | Dispersion Characteristics  | Shape Characteristics |
|-----------------------------------|-----------------------------|------------------------|
| Arithmetic mean                | Range                     | Skewness             |
| Weighted arithmetic mean       | Variance and standard deviation | Kurtosis       |
| Median                         |                             |                        |
| Quartiles                      |                             |                        |


```python
# Importing the necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating the dataset
np.random.seed(0)
data = {
    'Age': np.random.normal(loc=30, scale=5, size=1000).astype(int),  # Normally distributed data as integers
    'Salary': np.random.normal(loc=60000, scale=15000, size=1000).astype(int),  # Normally distributed data as integers
    'Years of Experience': np.random.exponential(scale=5, size=1000).astype(int)  # Exponentially distributed data as integers
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Displaying the DataFrame
df
```

## Central Tendency Characteristics

### Arithmetic Mean
The arithmetic mean $\mu$ is the sum of the values of the variable $j$ for all individuals $i, i = 1, ..., N$.

$\mu_j = \frac{1}{N} \sum_{i=1}^{N} x^j_i$

The arithmetic mean is sensitive to outliers.

```python
# Arithmetic Mean
mean_age = df['Age'].mean()
mean_salary = df['Salary'].mean()
mean_experience = df['Years of Experience'].mean()

print('Mean age:', mean_age)
print('Mean salary:', mean_salary)
print('Mean experience:', mean_experience)
```

### Weighted Arithmetic Mean
When variables do not have the same importance, a weight is assigned to each of them. In this case, the weighted arithmetic mean is calculated as follows:

$\lambda_j = \frac{\sum_{i=1}^{N} (w_i \cdot x^j_i)}{\sum_{i=1}^{N} (w_i)}$

```python
# Weighted Arithmetic Mean (Example with arbitrary weights)
weights = np.random.rand(1000)
weighted_mean_age = np.average(df['Age'], weights=weights)
weighted_mean_salary = np.average(df['Salary'], weights=weights)
weighted_mean_experience = np.average(df['Years of Experience'], weights=weights)

print('Weighted mean age: ', weighted_mean_age)
print('Weighted mean salary: ', weighted_mean_salary)
print('Weighted mean experience: ', weighted_mean_experience)
```

### Median
Given a set of $N$ data points sorted in ascending order, the median is the value of the variable that divides the dataset into two equal parts.
- If $N$ is odd $N = 2n + 1$, the median is the data point at position $n$.
- If $N$ is even $N = 2n$, the median is the data point at position $n$ or $n + 1$, or the average of the two.

In the case of a normal distribution, the median and the mean are equal.

```python
# Median
median_age = df['Age'].median()
median_salary = df['Salary'].median()
median_experience = df['Years of Experience'].median()

print('Median age:', median_age)
print('Median salary:', median_salary)
print('Median experience:', median_experience)
```

### Quartiles
Given a set of $N$ data points sorted in ascending order, the quartiles are the values $Q1$, $Q2$, and $Q3$ that divide the dataset into four equal parts.
- The first quartile $Q1$ is the smallest data point such that at least one-quarter of the data points are less than or equal to $Q1$.
- The third quartile $Q3$ is the smallest data point such that at least three-quarters of the data points are less than or equal to $Q3$.
- The second quartile $Q2$ corresponds to the median.

```python
# Quartiles
quartiles_age = df['Age'].quantile([0.25, 0.5, 0.75])
quartiles_salary = df['Salary'].quantile([0.25, 0.5, 0.75])
quartiles_experience = df['Years of Experience'].quantile([0.25, 0.5, 0.75])

print('Quartiles age:\n', quartiles_age)
print('Quartiles salary:\n', quartiles_salary)
print('Quartiles experience:\n', quartiles_experience)
```

## Dispersion Characteristics

### Range
The range is the difference between the largest and smallest values. This characteristic depends on outliers and is therefore not very reliable.

```python
# Range

range_age = df['Age'].max() - df['Age'].min()
range_salary = df['Salary'].max() - df['Salary'].min()
range_experience = df['Years of Experience'].max() - df['Years of Experience'].min()

print('Range age:', range_age)
print('Range salary:', range_salary)
print('Range experience:', range_experience)
```

### Variance and Standard Deviation
Variance is the average of the squared deviations from the mean.

$\text{var}(x) = \rho^2_x = \frac{1}{N} \sum_{i=1}^{N} ( x_i - \mu)^2 $

where $\mu$ is the arithmetic mean of the dataset.

The standard deviation $\sigma_x$ is the positive square root of the variance:

$\rho_x = \sqrt{\frac{1}{N} \sum_{i=1}^{N} ( x_i - \mu)^2}$

```python
# Variance

variance_age = df['Age'].var()
variance_salary = df['Salary'].var()
variance_experience = df['Years of Experience'].var()

print('Variance age:', variance_age)
print('Variance salary:', variance_salary)
print('Variance experience:', variance_experience)
```

```python
# Standard Deviation

std_dev_age = df['Age'].std()
std_dev_salary = df['Salary'].std()
std_dev_experience = df['Years of Experience'].std()

print('STD Deviation age:', std_dev_age)
print('STD Deviation salary:', std_dev_salary)
print('STD Deviation experience:', std_dev_experience)
```

## Shape Characteristics

### Skewness Coefficient
The skewness coefficient corresponds to the third moment of the centered reduced variable. For a uniform distribution $x$, the skewness coefficient is given by the formula:

$\gamma_1 = E \left[ \left( \frac{x - \mu}{\rho} \right)^3 \right]$

where $E$ denotes the expectation of $x$, $\mu$ the mean, and $\sigma $ the standard deviation. Alternatively,

$\gamma_1 = \frac{\mu_3}{\mu_2^{3/2}}$

where $\mu_i$ are the centered moments of order $i$.

There are three cases:
- $\gamma_1 < 0$ if the distribution of the variable is spread to the left (the tail is on the left) and the mean is less than the median.
- $\gamma_1 = 0$ if the distribution of the variable is symmetric.
- $\gamma_1 > 0$ if the distribution of the variable is spread to the right (the tail is on the right) and the mean is greater than the median.

```python
# Skewness

skewness_age = df['Age'].skew()
skewness_salary = df['Salary'].skew()
skewness_experience = df['Years of Experience'].skew()

print('Skewness age:', skewness_age)
print('Skewness salary:', skewness_salary)
print('Skewness experience:', skewness_experience)
```

```python
# Plotting the Skewness

plt.figure(figsize=(15, 5))

# Age Skewness
plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_age + 1, max(np.histogram(df['Age'], bins=20)[0]) / 2, f'Skewness: {skewness_age:.2f}', color='red')

# Salary Skewness
plt.subplot(1, 3, 2)
plt.hist(df['Salary'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_salary + 2000, max(np.histogram(df['Salary'], bins=20)[0]) / 2, f'Skewness: {skewness_salary:.2f}', color='red')

# Years of Experience Skewness
plt.subplot(1, 3, 3)
plt.hist(df['Years of Experience'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Years of Experience Distribution')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.axvline(mean_experience, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_experience + 0.5, max(np.histogram(df['Years of Experience'], bins=20)[0]) / 2, f'Skewness: {skewness_experience:.2f}', color='red')

plt.tight_layout()
plt.show()
```

### Kurtosis Coefficient
The kurtosis coefficient corresponds to the ratio of the fourth moment of the centered reduced variable to the fourth power of the standard deviation. It is given by the following equation:

$\gamma_2 = E \left[ \left( \frac{X - \mu}{\sigma_x} \right)^4 \right] $

- If $\gamma_2 = 0$, the distribution is called mesokurtic.
- If $\gamma_2 > 0$, the distribution is more concentrated than the normal; it is called leptokurtic. In finance, this coefficient is used to determine more frequent abnormal values.
- If $\gamma_2 < 0$, the distribution is flatter than the normal; it is called platykurtic. In finance, this coefficient is used to determine more frequent abnormal values.

```python
# Kurtosis

kurtosis_age = df['Age'].kurtosis()
kurtosis_salary = df['Salary'].kurtosis()
kurtosis_experience = df['Years of Experience'].kurtosis()

print('Kurtosis age:', kurtosis_age)
print('Kurtosis salary:', kurtosis_salary)
print('Kurtosis experience:', kurtosis_experience)
```

```python
# Plotting the Kurtosis

plt.figure(figsize=(15, 5))

# Age Kurtosis
plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_age + 1, max(np.histogram(df['Age'], bins=20)[0]) / 2, f'Kurtosis: {kurtosis_age:.2f}', color='red')

# Salary Kurtosis
plt.subplot(1, 3, 2)
plt.hist(df['Salary'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.axvline(mean_salary, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_salary + 2000, max(np.histogram(df['Salary'], bins=20)[0]) / 2, f'Kurtosis: {kurtosis_salary:.2f}', color='red')

# Years of Experience Kurtosis
plt.subplot(1, 3, 3)
plt.hist(df['Years of Experience'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Years of Experience Distribution')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.axvline(mean_experience, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_experience + 0.5, max(np.histogram(df['Years of Experience'], bins=20)[0]) / 2, f'Kurtosis: {kurtosis_experience:.2f}', color='red')

plt.tight_layout()
plt.show()
```

## 4.2 Graphical Representation of Quantitative Data

### Histogram

An histogram is a chart that represents the distribution of a quantitative statistical variable using adjacent bars. To plot a histogram for continuous variables, we group them into classes.

```python
# Plotting the Histogram

plt.figure(figsize=(15, 5))

# Age Histogram
plt.subplot(1, 3, 1)
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Salary Histogram
plt.subplot(1, 3, 2)
plt.hist(df['Salary'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Years of Experience Histogram
plt.subplot(1, 3, 3)
plt.hist(df['Years of Experience'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Years of Experience Distribution')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Box Plot

A box plot, also known as a whisker plot, is a graphical representation of the statistical characteristics of a dataset, including the median (Q2), quartiles (Q1, Q2, and Q3), and range.
- The limits of the rectangle correspond to the first and third quartile.
- The limits of the whiskers correspond to the minimum and maximum values of the dataset.

```python
# Plotting the Box Plot

plt.figure(figsize=(15, 5))

# Age Box Plot
plt.subplot(1, 3, 1)
plt.boxplot(df['Age'], vert=True, patch_artist=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Values')

# Salary Box Plot
plt.subplot(1, 3, 2)
plt.boxplot(df['Salary'], vert=True, patch_artist=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Values')

# Years of Experience Box Plot
plt.subplot(1, 3, 3)
plt.boxplot(df['Years of Experience'], vert=True, patch_artist=True)
plt.title('Years of Experience Distribution')
plt.xlabel('Years of Experience')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```
