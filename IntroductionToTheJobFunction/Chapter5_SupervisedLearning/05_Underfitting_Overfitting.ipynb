# 5. Underfitting and Overfitting

In this notebook, we will explain the concepts of underfitting and overfitting in machine learning. We will then demonstrate these concepts with examples using regression models.

---

## 5.1 Underfitting

**Underfitting** occurs when a model is too simple to capture the underlying patterns in the data. It performs poorly on both the training data and unseen data because it fails to capture the complexity of the data.

### Characteristics of Underfitting
- High bias
- Low variance
- Poor performance on training data
- Poor performance on testing data

### Example of Underfitting

We will create a linear regression model on a non-linear dataset to demonstrate underfitting.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate some non-linear data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Feature: 100 random numbers between 0 and 10
y = 15 * X**2 + np.random.randn(100, 1) * 50  # Target: quadratic relationship with some noise

# Create and train a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict using the linear model
y_pred = model.predict(X)

# Plot the data and the linear regression model
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Underfitting Example: Linear Regression on Non-linear Data')
plt.legend()
plt.show()

# Calculate and print the MSE
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error (MSE) for Underfitting Example:", mse)
```

## 5.2 Overfitting

**Overfitting** occurs when a model is too complex and captures the noise in the training data rather than the underlying patterns. It performs well on the training data but poorly on unseen data because it has essentially memorized the training data.

### Characteristics of Overfitting
    Low bias
    High variance
    Excellent performance on training data
    Poor performance on testing data
    
### Example of Overfitting
We will create a high-degree polynomial regression model on the same non-linear dataset to demonstrate overfitting.

```python
# Import necessary libraries
from sklearn.preprocessing import PolynomialFeatures

# Generate some non-linear data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Feature: 100 random numbers between 0 and 10
y = 15 * X**2 + np.random.randn(100, 1) * 50  # Target: quadratic relationship with more noise

# Create polynomial features (degree 40) to induce overfitting
poly_features = PolynomialFeatures(degree=40, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Create and train a polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Predict using the polynomial model
y_poly_pred = poly_model.predict(X_poly)

# Plot the data and the polynomial regression model
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data')
X_sorted, y_poly_pred_sorted = zip(*sorted(zip(X, y_poly_pred)))
plt.plot(X_sorted, y_poly_pred_sorted, color='red', linewidth=2, label='Polynomial Regression (Degree 40)')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Overfitting Example: Polynomial Regression (Degree 40)')
plt.legend()
plt.show()

# Calculate and print the MSE
poly_mse = mean_squared_error(y, y_poly_pred)
print("Mean Squared Error (MSE) for Overfitting Example:", poly_mse)
```

---

By examining the Mean Squared Error (MSE) and the visualizations of these regression models, we can clearly see the effects of underfitting and overfitting. Underfitting results in a model that is too simple and cannot capture the complexity of the data, while overfitting results in a model that is too complex and captures noise in the data.

In practice, the goal is to find a balance between underfitting and overfitting, which is often referred to as the bias-variance tradeoff. This balance ensures that the model generalizes well to unseen data.
