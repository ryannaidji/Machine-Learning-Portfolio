## Bar Plot

- The **bar plot** is used to represent the counts of different categories of a qualitative variable.
- The bar plot can be represented horizontally or vertically.
- In some cases, the bar plot can also be used to summarize numeric data via the mean.

### Example of a Bar Plot in Python


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {'Category': ['A', 'B', 'C', 'D'],
        'Value': [10, 20, 15, 25]}

df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=df)
plt.title('Bar Plot of Values by Category')
plt.show()
```

## Lollipop Chart

- The **lollipop chart** is an attractive diagram due to its simplicity.
- Like the bar plot, it is used to represent the counts of different categories of a qualitative variable.

### Example of a Lollipop Chart in Python


```python
# Create the lollipop chart
plt.figure(figsize=(8, 6))
plt.stem(df['Category'], df['Value'], basefmt=" ")
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Lollipop Chart of Values by Category')
plt.show()
```

## Parallel Coordinates

- Parallel coordinates allow representing an observation of dimension n by n parallel axes on a plane, each axis representing a dimension.
- The interest of parallel coordinates lies in the fact that certain geometric properties of the variables can be easily transformed and interpreted in 2D.

### Example of Parallel Coordinates in Python


```python
from pandas.plotting import parallel_coordinates

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Feature1': [1, 2, 3, 4],
    'Feature2': [4, 3, 2, 1],
    'Feature3': [2, 3, 4, 5]
}

df = pd.DataFrame(data)

# Create the parallel coordinates plot
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'Category', color=('#556270', '#4ECDC4', '#C7F464', '#FF6B6B'))
plt.title('Parallel Coordinates Plot')
plt.show()
```

## Word Cloud

- The **word cloud** is a visual representation of the most used keywords in a text.
- Keywords are displayed in font sizes and thicknesses proportional to their frequency.

### Example of a Word Cloud in Python


```python
from wordcloud import WordCloud

# Sample text
text = "data visualization matplotlib seaborn python data analysis visualization matplotlib seaborn python"

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()
```

## Radar Chart

- The **radar chart** (also known as spider chart, star chart) is a two-dimensional chart designed to plot one or more series of values on multiple quantitative variables.
- It is formed by as many axes as there are variables. The values (or series) are displayed inside the web.
- Values must be normalized to be overlaid.

### Example of a Radar Chart in Python


```python
import numpy as np

# Sample data
labels = ['Feature1', 'Feature2', 'Feature3', 'Feature4']
values = [4, 3, 2, 5]

# Number of variables
num_vars = len(labels)

# Compute angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is made in a circular (360 degree) space, so we need to "complete the loop" and append the start value to the end.
values += values[:1]
angles += angles[:1]

# Plotting
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title('Radar Chart')
plt.show()
```
