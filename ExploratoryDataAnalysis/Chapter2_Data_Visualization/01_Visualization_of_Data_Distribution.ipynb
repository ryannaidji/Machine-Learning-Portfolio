# Visualization Techniques

Data visualization makes analysis much easier and faster, hence the need to choose appropriate visualization tools.

- Libraries like **Matplotlib** and **Seaborn** are very useful and effective for data visualization.

## Visualization of Distribution

### Box Plot

- The **box plot** is one of the most commonly used charts.
- It summarizes 5 important parameters of the variable: the minimum value, the first quartile $Q1$, the median (or second quartile $Q2$, the third quartile $Q3$, and the maximum value.
- The line dividing the box into two parts represents the median.
- The limits of the box show the upper $Q3$ and lower $Q1$ quartiles.
- The whiskers show the highest and lowest values of the variable, excluding outliers.

### Example of a Box Plot in Python

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Category': ['A', 'A', 'A', 'A', 'B', 'B', 'B', 'B'],
        'Value': [1, 2, 2, 3, 4, 5, 5, 6]}

df = pd.DataFrame(data)

# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Box Plot of Values by Category')
plt.show()
```

### Histogram

- An **histogram** represents an estimate of the density of a quantitative variable.
- The shape of the histogram is obtained by distributing the data according to a set of intervals. Thus, it can be different depending on the number of defined intervals.
- Identify the minimum (min) and maximum (max) values of the variable to be explored.
- Divide the interval \([min; max]\) into I sub-intervals.
- Count the observations for which the value of the variable falls within each interval (also called classes).
- Represent the number of observations per interval with a bar whose area is proportional to the counts.

### Example of an Histogram in Python


```python
import numpy as np

# Sample data
np.random.seed(42)
data = np.random.randn(1000)

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### How Many Bins?

- The number of bins in a histogram has a significant impact on the visualization of the actual distribution of a numeric variable.
- There are two commonly used methods to calculate the optimal number of bins:
  - Method 1: Sturges' Rule
  - Method 2: Freedman-Diaconis Rule

#### Sturges' Rule

- A mathematical formula proposed by Herbert Sturges to divide a range of values into slices for statistical description.
- For a sample of N observed values, the approximate value for the number of bins \(k\) based on the sample size $N$:

$$k = 1 + \text{ceil}\lceil \log_2 N \rceil$$

where $\log_2$ is the logarithm in base 2.

#### Freedman-Diaconis Rule

- Takes into account not only the sample size $N$ but also its dispersion.

$$I_w = \frac{2(Q_3 - Q_1)}{\sqrt[3]N}$$
$$k = \text{ceil}\lceil \frac{(\max(x) - \min(x))}{I_w} \rceil$$

where $Q3$ and $Q3$ are the quartiles.
$N$ is the sample size.
$x$ is the variable to be studied.

### Symmetric/Asymmetric Histogram

- A histogram is symmetric when its profile on the left is identical or very similar to its profile on the right around a mode.
- Data skewed to the left (also called negatively skewed data) are named because the "tail" of the distribution points to the left.
- Data skewed to the right (also called positively skewed data) are named because the "tail" of the distribution points to the right.

### Multimodal Histogram

- The modes of a histogram correspond to the most abundant classes (intervals) locally.
- A multimodal histogram has several modes (2, 3, ...).
- Often reflects a population composed of several distinct sub-populations.

### Density Plot

- A **density plot** is also a graphical representation of the distribution of a numeric variable.
- It presents a smoothed histogram.
- The transition from a histogram to a density curve involves smoothing the peaks more or less strongly. This is often done using estimation techniques.

### Example of a Density Plot in Python


```python
# Create the density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data, fill=True)
plt.title('Density Plot of Random Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

### Violin Plot

- A **violin plot** allows visualizing the distribution of a numeric variable.
- It consists of two mirrored density plots and a box plot, providing a deeper understanding of the density.

### Example of a Violin Plot in Python


```python
# Create the violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Violin Plot of Values by Category')
plt.show()
```
