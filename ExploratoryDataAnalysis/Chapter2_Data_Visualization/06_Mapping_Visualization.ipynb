## Mapping Visualization

### Background Map
The first step in any geo-spatial data analysis is to draw the background map of the area of interest. A map is essentially a set of polygons displayed on a two-dimensional canvas and characterized by their coordinates. We can use the most popular tool in cartography, `geopandas`.

### Choropleth Map
A choropleth map displays divided geographical areas or regions that are colored in relation to a numeric variable. To build a choropleth map, we need two data inputs:
1. A set of geographical regions and their boundary coordinates.
2. A numeric value for each region, used for the color.

We can use the most popular tool in cartography, `geoplot`.


```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the shapefile
shapefile_path = "./ne_110m_admin_0_countries//ne_110m_admin_0_countries.shp"
world = gpd.read_file(shapefile_path)

# Plot the background map
plt.figure(figsize=(15, 10))
world.plot(ax=plt.gca(), color='lightgray', edgecolor='black')
plt.title('Background Map')
plt.show()
```

```python
import geopandas as gpd
import geoplot as gplt
import matplotlib.pyplot as plt
import numpy as np

# Load the shapefile
shapefile_path = "./ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"
world = gpd.read_file(shapefile_path)

# Create a sample numeric value for each region (e.g., random population data)
world['sample_data'] = np.random.rand(len(world)) * 100

# Ensure MultiPolygon geometries are handled correctly
world = world.explode(index_parts=False)

# Plot the choropleth map
plt.figure(figsize=(15, 10))
gplt.choropleth(
    world,
    hue='sample_data',
    cmap='coolwarm',
    linewidth=0.5,
    edgecolor='white',
    legend=True
)
plt.title('Choropleth Map')
plt.show()
```
