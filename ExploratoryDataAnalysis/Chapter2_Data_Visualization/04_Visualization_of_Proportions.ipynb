## Visualization of Proportions

### Grouped or Stacked Bar Chart
Grouped or stacked bar charts display the relative size (in the form of a total number, percentage, or other metric) of a categorical variable, subdivided by color based on a subgroup.

### Pie Chart / Doughnut Chart
Pie charts, also known as camembert or tart charts, present proportions of qualitative (categorical) data. The size of each slice represents the proportion of each category. The doughnut chart follows the same principle as the pie chart.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for grouped bar chart
categories = ['Category 1', 'Category 2', 'Category 3']
subgroups = ['Subgroup A', 'Subgroup B']
data = {
    'Category 1': [5, 7],
    'Category 2': [8, 6],
    'Category 3': [6, 9]
}

df_grouped_bar = pd.DataFrame(data, index=subgroups)

# Plot grouped bar chart
ax = df_grouped_bar.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Grouped Bar Chart')
plt.xlabel('Subgroups')
plt.ylabel('Values')
plt.legend(title='Categories')
plt.show()
```

```python
# Plot stacked bar chart
ax = df_grouped_bar.plot(kind='bar', figsize=(10, 6), rot=0, stacked=True)
plt.title('Stacked Bar Chart')
plt.xlabel('Subgroups')
plt.ylabel('Values')
plt.legend(title='Categories')
plt.show()
```

```python
# Sample data for pie chart
labels = ['Category A', 'Category B', 'Category C']
sizes = [15, 30, 45]

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# Plot doughnut chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, wedgeprops={'width': 0.4})
plt.title('Doughnut Chart')
plt.show()
```
