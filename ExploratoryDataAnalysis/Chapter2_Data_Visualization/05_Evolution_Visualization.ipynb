## Evolution Visualization

### Line Charts
Line charts display information as a series of data points connected by straight lines. Categories appear along the x-axis and statistics along the y-axis. Time series charts can be considered a special case of line charts.

### Time Series Chart
A time series chart visualizes trends of totals or numerical values over time. Time series charts allow us to extract knowledge about the trend of a variable's evolution over time.

### Area Chart
An area chart represents the evolution of a quantitative variable according to another quantitative variable. The area between the axis and the line is generally emphasized by colors, textures, or hatching. Typically, an area chart compares at least two quantities.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for line chart
x = np.arange(1, 11)
y1 = np.random.randint(10, 20, size=10)
y2 = np.random.randint(5, 15, size=10)

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='o', label='Series 1')
plt.plot(x, y2, marker='o', label='Series 2')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
# Sample data for time series chart
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100).cumsum()

# Create DataFrame
df_time_series = pd.DataFrame({'Date': dates, 'Value': values})
df_time_series.set_index('Date', inplace=True)

# Plot time series chart
plt.figure(figsize=(10, 6))
df_time_series.plot(ax=plt.gca())
plt.title('Time Series Chart')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

```python
# Sample data for area chart
x = np.arange(1, 11)
y1 = np.random.randint(10, 20, size=10)
y2 = np.random.randint(5, 15, size=10)

# Plot area chart
plt.figure(figsize=(10, 6))
plt.fill_between(x, y1, color="skyblue", alpha=0.4, label='Series 1')
plt.fill_between(x, y2, color="sandybrown", alpha=0.4, label='Series 2')
plt.plot(x, y1, color="Slateblue", alpha=0.6, linewidth=2)
plt.plot(x, y2, color="Sienna", alpha=0.6, linewidth=2)
plt.title('Area Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
