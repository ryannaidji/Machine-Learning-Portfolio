# Visualization of Data Correlation

## Scatter Plot

- A **scatter plot** allows representing one numerical variable against another numerical variable: Y ~ X
- Each point represents an observation.
- The positions on the X (horizontal) and Y (vertical) axes represent the values of the two variables.

### Example of a Scatter Plot in Python


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.randn(100) * 0.1

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Y ~ X')
plt.show()
```

## Connected Scatter Plot

- A **connected scatter plot** is very similar to a scatter plot, except that the points are connected by lines.
- This means that the values of the observations on the X-axis are ordered for this type of representation to be useful.
- Connected scatter plots are often used for time series where the X-axis represents time.

### Example of a Connected Scatter Plot in Python


```python
# Sample data for connected scatter plot
time = np.arange(0, 100)
values = np.sin(time / 10) + np.random.randn(100) * 0.1

# Create a connected scatter plot
plt.figure(figsize=(10, 6))
plt.plot(time, values, marker='o')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Connected Scatter Plot of Values Over Time')
plt.show()
```

## Heatmap

- A **heatmap** is a graphical representation of data where individual values contained in a matrix are represented as colors.
- It allows displaying an overall view of numerical data.

### Example of a Heatmap in Python


```python
# Sample data for heatmap
data = np.random.rand(10, 12)
df_heatmap = pd.DataFrame(data, columns=[f'Col_{i}' for i in range(1, 13)])

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, annot=True, cmap='coolwarm')
plt.title('Heatmap of Random Data')
plt.show()
```

## Bubble Plot

- A **bubble plot** is a scatter plot where additional dimensions are added for more information.
- It requires 3 numerical variables as input: one is represented by the X-axis, one by the Y-axis, and one by the size of the bubbles.
- The surface area of the bubbles should be proportional to the value of the data.

### Example of a Bubble Plot in Python


```python
# Sample data for bubble plot
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.randn(100) * 0.1
size = np.random.rand(100) * 1000

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=size, alpha=0.5, edgecolors="w")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bubble Plot of Y ~ X with Bubble Size')
plt.show()
```

## 2D Density Plot

- A **2D density plot** or 2D histogram is an extension of the 1D histogram.
- It shows the distribution of values of a dataset over the range of two quantitative variables.

### Example of a 2D Density Plot in Python


```python
# Sample data for 2D density plot
x = np.random.randn(1000)
y = np.random.randn(1000)

# Create a 2D density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(x=x, y=y, cmap='Blues', fill=True, bw_adjust=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Density Plot')
plt.show()
```

## Correlogram

- A **correlogram** allows analyzing the relationship between each pair of numerical variables in a matrix.
- The correlation between each pair of variables is visualized by a scatter plot or a symbol representing the correlation (bubble, line, number, etc.).
- The diagonal represents the distribution of each variable, using a histogram or density plot.

### Example of a Correlogram in Python


```python
# Sample data for correlogram
df_correlogram = pd.DataFrame({
    'A': np.random.randn(100),
    'B': np.random.randn(100) * 2,
    'C': np.random.randn(100) / 2,
    'D': np.random.randn(100) + 1
})

# Create a correlogram
sns.pairplot(df_correlogram)
plt.suptitle('Correlogram of Sample Data', y=1.02)
plt.tight_layout()
plt.show()
```
